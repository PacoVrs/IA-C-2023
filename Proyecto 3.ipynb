```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load data from CSV
df = pd.read_csv('D:/ia/resultado.csv', header=None, names=['x1', 'y1', 'x2', 'y2', 'target1', 'target2'], sep=' ', dtype=float)

# Create the figure
fig = plt.figure()

# Add a 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Add the values when targets are 0
sc = ax.scatter(df[df['target1'] == 0]['x1'], df[df['target1'] == 0]['y1'], df[df['target1'] == 0]['target1'], c='blue', marker='o', label='Jump = 0')
sc = ax.scatter(df[df['target2'] == 0]['x2'], df[df['target2'] == 0]['y2'], df[df['target2'] == 0]['target2'], c='green', marker='x', label='Desp = 0')

# Add the values when targets are 1
sc = ax.scatter(df[df['target1'] == 1]['x1'], df[df['target1'] == 1]['y1'], df[df['target1'] == 1]['target1'], c='red', marker='o', label='Jump = 1')
sc = ax.scatter(df[df['target2'] == 1]['x2'], df[df['target2'] == 1]['y2'], df[df['target2'] == 1]['target2'], c='purple', marker='x', label='Desp = 1')

# Set labels for the axes
ax.set_xlabel('x1')
ax.set_ylabel('y1')
ax.set_zlabel('target')

# Show labels
ax.legend()

# Show the plot
plt.show()
```
